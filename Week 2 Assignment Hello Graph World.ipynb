In this assignment I will create the nodes and edges for an organizational chart as a simple "beachhead" program using NetworkX.

```python
# Import Libraries
import networkx as nx
import matplotlib.pyplot as plt
     
```

```python
G = nx.Graph()

# Kite Nodes

nodes = ["Jen", "Toni", "Constance", "Kanna", "Carlotta", "Chad", "Claudia", "Brandon", "Freddie","Dawn","Omkar"]
    
G.add_nodes_from(nodes)

# Kite Edges

edges = [
    ("Toni", "Jen"),
    ("Jen", "Constance"), 
    ("Jen", "Kanna"), 
    ("Jen", "Carlotta"), 
    ("Constance", "Chad"), 
    ("Constance", "Freddie"),
    ("Kanna", "Brandon"),
    ("Kanna", "Claudia"),
    ("Carlotta", "Dawn"),
    ("Carlotta","Omkar"),]

G.add_edges_from(edges)
```

Here I am initilizing the graph and creating the lists of nodes and edges to show the people and their relationships inside the organization and who reports directly to who.

```python
# Assign roles
roles = {
    "Jen": "Director",
    "Toni": "Senior Director",
    "Constance": "Manager",
    "Kanna": "Manager",
    "Carlotta": "Manager",
    "Brandon": "Analyst",
    "Chad": "Analyst",
    "Claudia": "Analyst",
    "Freddie": "Analyst",
    "Dawn": "Analyst",
    "Omkar": "Analyst"
}

nx.set_node_attributes(G, roles, "role")

# Create labels
labels = {node: f"{node}\n({G.nodes[node]['role']})" for node in G.nodes()}
```

To label each node with the title of each employee in the organization I assigned roles as attributes to each node and created labels to show both the name and the role of each node.

```python
# Draw the graph
plt.figure(figsize=(12,10))
nx.draw(G, labels=labels, with_labels=True,
        node_color="lightblue", node_size=2000, font_size=12)
plt.title("Org Chart with Role Labels", size=16)
plt.show()

```

When plotting the resulting org chart clearly lays out the relationships of each employee with their titles.
